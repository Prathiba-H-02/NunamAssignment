```python
import numpy as np
from scipy.signal import butter,filtfilt
# Filter requirements.
T = 60.0        # Sample Period
fs = 6944       # sample rate, Hz
cutoff = 1.5     # desired cutoff frequency of the filter, Hz ,      slightly higher than actual 1.2 Hz
nyq = 0.5 * fs  # Nyquist Frequency
n = len(df)
order=2
```

```python
def butter_lowpass_filter(data, cutoff, fs, order):
    normal = cutoff / nyq 
    b, a = butter(order, normal, btype='low', analog=False)
    y = filtfilt(b, a, data)
    return y
```

```python
out=butter_lowpass_filter(df['Unnamed: 0'],cutoff,fs,order)
```

```python
import plotly.graph_objs as go
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
            out = df['Unnamed: 0'],
            line =  dict(shape =  'spline' ),
            ))
fig.add_trace(go.Scatter(
            out = out,
            line =  dict(shape =  'spline' ),
            name = 'filtered signal'
            ))
fig.show()
```
